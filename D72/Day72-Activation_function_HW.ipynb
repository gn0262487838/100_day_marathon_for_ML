# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def ReLU(x):

    if x > 0:
        return x
    else:
        return 0

def dReLU(x):

    if x > 0:
        return 1
    else:
        return 0
```

```python
x = np.linspace(-1,1)
y = [ReLU(i) for i in x]

plt.figure()
plt.plot(x, y)
plt.title("ReLU:")
plt.show()
```

```python
z = [dReLU(i) for i in x]

plt.figure()
plt.plot(x, z)
plt.title("dReLU:")
plt.show()
```
