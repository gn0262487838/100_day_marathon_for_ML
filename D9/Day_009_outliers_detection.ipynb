# 檢視 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = '../D4/data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 開始檢視不同欄位是否有異常值

```python
# DAYS_BIRTH: 客戶申請貸款時的年齡
print(app_train['DAYS_BIRTH'])
print("=====我是分隔線=====")
print((app_train['DAYS_BIRTH'] / (-365)).describe())
```

```python
# DAYS_EMPLOYED: 申請貸款前，申請人已在現職工作的時間
print(app_train['DAYS_EMPLOYED'])
print("=====分隔線=====")
print((app_train['DAYS_EMPLOYED'] / 365).describe())
print("=====分隔線=====")
plt.figure(figsize=(8,8))
plt.title("DAYS_EMPLOYED Only On Application")
plt.xlabel("Years")
plt.ylabel("Counts")
plt.hist(app_train['DAYS_EMPLOYED'], color="orange",edgecolor="yellow",label="DAYS_EMPLOYED")
plt.grid()
plt.legend()
plt.show()
print(app_train['DAYS_EMPLOYED'].value_counts())
```

### 從上面的圖與數值可以看出, 365243 顯然是個奇怪的數值

```python
anom = app_train[app_train['DAYS_EMPLOYED'] == 365243]
non_anom = app_train[app_train['DAYS_EMPLOYED'] != 365243]
print('The non-anomalies default on %.2f %% of loans' % (100 * non_anom['TARGET'].mean()))
print('The anomalies default on %.2f %% of loans' % (100 * anom['TARGET'].mean()))
print('There are %s anomalous days of employment' % len(anom))
```

```python
print("異常值365243佔全體資料約 {:.2f} %".format((sum(app_train['DAYS_EMPLOYED'] == 365243)/len(app_train)) * 100))
```

```python
# 新增一個欄位: DAYS_EMPLOYED_ANOM 來標記 DAYS_EMPLOYED 是否異常
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
print(app_train['DAYS_EMPLOYED_ANOM'].value_counts())

# 這邊我們用 nan 將異常值取代，注意這裡的取代方式喔!!!
replaceDict = {365243: np.nan}
app_train['DAYS_EMPLOYED'].replace(replaceDict, inplace = True)

# way1:
# plt.figure(figsize=(8,8))
# plt.title("Days Employment Histogram")
# plt.xlabel("Days Employment")
# plt.ylabel("Counts")
# plt.hist(app_train['DAYS_EMPLOYED'], color="orange",label="Days Employment")
# plt.grid()
# plt.legend()
# plt.show()

# way2:
app_train['DAYS_EMPLOYED'].plot.hist(title = 'Days Employment Histogram');
plt.xlabel('Days Employment');
```

```python
# 檢查 OWN_CAR_AGE: 貸款人的車齡
plt.figure(figsize=(8,8))
plt.title("Own Car Age")
plt.xlabel("Car Age")
plt.ylabel("Counts")
plt.hist(app_train[~app_train.OWN_CAR_AGE.isnull()]['OWN_CAR_AGE'], color="orange",edgecolor="black", label="Car Age")
plt.grid()
plt.legend()
plt.show()
app_train['OWN_CAR_AGE'].value_counts()
```
