```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = '../D7/data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
df
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'], y=train_Y, line_kws={"color":"orange"}, scatter_kws={"color":"black"})
plt.show()

# 做線性迴歸, 觀察分數
MMEncoder = MinMaxScaler()
estimator = LinearRegression()
train_X = MMEncoder.fit_transform(df)
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
"""
Your Code Here
"""
df['1stFlrSF'] = df['1stFlrSF'].clip(lower=400,upper=2500)
sns.regplot(x=df['1stFlrSF'], y=train_Y, line_kws={"color":"orange"}, scatter_kws={"color":"black"})
plt.show()

# 做線性迴歸, 觀察分數
MMEncoder = MinMaxScaler()
estimator = LinearRegression()
train_X = MMEncoder.fit_transform(df)
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
Q:請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

A:
原先離群值超過2500過高，導致線性回歸分數比較低，但在限縮邊界為2500後從83多提高至88多，有顯著的進步。
但捨去離群值後的回歸分數只從88增加到89，感覺上應沒有差別太大。
再次實驗捨去2250以上之值，回歸分數反而下降到88，故實質上極端值影響已不大。

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
"""
Your Code Here
"""
condition = (df['1stFlrSF'] > 400) & (df['1stFlrSF'] < 2250)
df = df[condition]
train_Y = train_Y[condition]
sns.regplot(x=df['1stFlrSF'], y=train_Y, line_kws={"color":"orange"}, scatter_kws={"color":"black"})
plt.show()

# 做線性迴歸, 觀察分數
MMEncoder = MinMaxScaler()
estimator = LinearRegression()
train_X = MMEncoder.fit_transform(df)
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```
