```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

%matplotlib inline
plt.style.use('ggplot')

# 忽略警告訊息
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = '../D4/data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""

unique_house_type = app_train["HOUSETYPE_MODE"].unique()
print("=====我是分隔線=====")
print("型態:", type(unique_house_type))
print(unique_house_type)
print("=====我是分隔線=====")
print("查看np.nan什麼型態: ", type(np.nan))
print("=====我是分隔線=====")

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(15,50))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    # 注意使用way1和way2畫出來的bins不一樣喔!!!
    # way1
    sns.distplot(app_train.loc[app_train["HOUSETYPE_MODE"] == unique_house_type[i], "AMT_CREDIT"], 
                hist=True, 
                 kde=False, 
                 rug=False,
                hist_kws={"histtype":"step", "color":"green", "linewidth":"3"})
    
    # Way2
    # app_train.loc[app_train["HOUSETYPE_MODE"] == unique_house_type[i], "AMT_CREDIT"].hist()
    
    plt.title(str(unique_house_type[i]))
plt.show()    
```
