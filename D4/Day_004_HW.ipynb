```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
print("rows number: {}".format(app_train.shape[0]))
print("columns number: {}".format(app_train.shape[1]))
print("data shape: {}".format(app_train.shape))
```

#### 列出所有欄位

```python
import random
idx = random.randint(0,app_train.shape[1])
print("所有欄位:\n", app_train.columns)
print(f"取一個欄位名稱來看看: {app_train.columns[idx]}")
```

#### 截取部分資料

```python
print(f"從頭開始數:\n", app_train.head(10))
print(f"從尾開始數:\n", app_train.tail(10))
print(f"隨便截一段:\n", app_train[10:20])
```

還有各種數之不盡的資料操作，重點還是取決於實務中遇到的狀況和你想問的問題，在馬拉松中我們也會陸續提到更多例子

```python
# 利用iloc擷取表格畫面
# loc, iloc差在能否用index取值
app_train.iloc[0:10, 0:3]
```

```python
print(f"target總數: {app_train['TARGET'].shape[0]}")
print(f"總共有幾個target=1: {app_train['TARGET'].sum()}")
print(f"總共有幾個target=0: {app_train['TARGET'].shape[0] - app_train['TARGET'].sum()}")
```
